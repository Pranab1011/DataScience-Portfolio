```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
women_degrees.head()
```

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

```python
fig = plt.figure(figsize=(18,3))
direc = ['right', 'left', 'top', 'bottom']
for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for d in direc:
        ax.spines[d].set_visible(False)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(12,24))
direc = ['right', 'left', 'top', 'bottom']
for sp in range(0,6):
    ax = fig.add_subplot(6,3,3*sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for d in direc:
        ax.spines[d].set_visible(False)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.set_yticks([0,50,100])
    ax.axhline(50,c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
        
    ax1 = fig.add_subplot(6,3,3*sp+3)
    ax1.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax1.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for d in direc:
        ax1.spines[d].set_visible(False)
    ax1.set_xlim(1968,2011)
    ax1.set_ylim(0,100)
    ax1.set_title(other_cats[sp])
    ax1.set_yticks([0,50,100])
    ax1.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax1.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')

ax.tick_params(labelbottom='on')
ax1.tick_params(labelbottom='on')

    
for sp in range(0,5):
    ax2 = fig.add_subplot(6,3,3*sp+2)
    ax2.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, linewidth=3)
    ax2.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, linewidth=3)
    ax2.set_title(lib_arts_cats[sp])
    for d in direc:
        ax2.spines[d].set_visible(False)
    ax2.set_xlim(1968,2011)
    ax2.set_ylim(0,100)
    ax2.set_yticks([0,50,100])
    ax2.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax2.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
ax2.tick_params(labelbottom='on')

plt.savefig('gender_gap.jpeg')
plt.show()
```
